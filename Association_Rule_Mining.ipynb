## Data preparation

### Subtask:
Prepare the data for association rule mining algorithms by converting it into a transaction format.


**Reasoning**:
Prepare the data for association rule mining by dropping the 'Unnamed: 0' column and converting boolean values to integers.



```python
df_cleaned = df.drop('Unnamed: 0', axis=1)
df_cleaned = df_cleaned.astype(int)
display(df_cleaned.head())
```

## Apply apriori algorithm

### Subtask:
Apply the Apriori algorithm to find frequent itemsets.


**Reasoning**:
Apply the Apriori algorithm to find frequent itemsets with a minimum support of 0.01.



```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_apriori = apriori(df_cleaned, min_support=0.01, use_colnames=True)
display(frequent_itemsets_apriori.head())
```

## Generate association rules (apriori)

### Subtask:
Generate association rules from the frequent itemsets obtained from Apriori.


**Reasoning**:
Generate association rules from the frequent itemsets using the association_rules function with lift as the metric and a minimum threshold of 1, and display the head of the resulting DataFrame.



```python
from mlxtend.frequent_patterns import association_rules

association_rules_apriori = association_rules(frequent_itemsets_apriori, metric="lift", min_threshold=1)
display(association_rules_apriori)
```
